```python
glob = 1

#exchange
def xchgScope(x):
    #global 키워드를 통해 위에 있는 glob을 현재 스코프로 변경
    global glob
    glob = 7
    return glob + x

print(glob)
print(xchgScope(3))
print(glob)
```

```python
class Person:
    name = "Default"
    
    #파이썬의 매서드에는 반드시 self를 첫번째 인자로 배치해줘야 한다. 
    def personPrint(self):
        print("내 이름은 {0}".format(self.name))
# init 즉 메모리 할당 작업을 담당 -> 다른 용어로 생성자라 부름
# 생성자는 클래스 이름과 동일하다. 
p1 = Person()
p1.personPrint()

p1.name = "홍길동"
p1.personPrint()
Person.personPrint(p1)
```

```python
class Test:
    test1 = 3
    _test2 = 7      #prtected : 가문의 가보
    __test3 = 33    #private : 나만의 사생활
    
    def testPrint(self):
        print(self.__test3)
    
test = Test()

print(test.test1)
print(test._test2)
test.testPrint()
```

```python
#생성자(Constructor)은 class 이름과 함수의 이름이 같다.
class Person:
    def __init__(self, name, age):
        print("생성자 호출!")
        self.name = name
        self.age = age
        
    def print(self):
        print("이름은 {0}, 나이는 {1}".format(self.name, self.age))
        
p = Person("고고싱", 77)
p.print()

print(p.__dict__) #__dict__ = 상태 정보 알려주는 것
```

```python
class Student(Person):
    def __init__(self, name, age, major):
        #위의 Person의 생성자 객체정보를 가지고 온다. 
        super(Student, self).__init__(name, age)
        self.major = major
        
    def printStudent(self):
        print("Student - name: {0}, age: {1}, major: {2}".format(self.name, self.age, self.major))
        
s = Student("고고싱", 77, "물리학")

print(s.__dict__)
```

```python
import time
import multiprocessing

def count_down(x):
    while True:
        if x == 0:
            break
            
        print("카운트 다운 ... %d" % x)
        x -= 1
        time.sleep(1)
        
# 프로세스 생성 -> 프로세스 담당 작업은 count_down 함수, 인자 = 5       
p = multiprocessing.Process(target=count_down, args=(5, ))
p.start()
```

```python
import threading
import os

def info():
    print('모듈명: ', __name__)
    print('프로세스 id: ', os.getpid())
    print('스레드 id: ', threading.get_ident())
    print('이런식으로 많이 집어넣으면 Context Switching도 확인할 수 있음')
    
def print_hello():
    print('안녕')
    info()
    time.sleep(1)
    print('잠자다 일어난다\n')
    
if __name__ == '__main__':
    # 현재 시간 기록
    start = time.perf_counter()
    threads = []
    
    # 스레드 5개 생성
    # 실제 컴퓨터 성능에 따라서 이 부분에 많이 혼성될 수도 있고 덜 혼성될 수도 있음
    # 컴퓨터가 압도적으로 좋거나 압도적으로 후지면 잘 혼성되지 않는다.
    for _ in range(100):
        #스레드 생성
        t = threading.Thread(target = print_hello)
        #준비
        t.start() 
        threads.append(t)
        
    # 스레드 구동
    for thread in threads:
        #동작
        thread.join()
        
    # 현재 시간 기록(현재 시간 - 이전 시간) -> 걸린 시간
    finish = time.perf_counter()
    
    print('종료: ', str(round(finish - start, 2)) + '초')
```

```python
import threading

x = 0

def increment_global():
    global x
    x += 1
    
def thread_processing():
    for _ in range(1000000):
        increment_global()
        
def thread_main():
    global x
    x = 0
    
    t1 = threading.Thread(target = thread_processing)
    t2 = threading.Thread(target = thread_processing)
    
    t1.start()
    t2.start()
    
    t1.join()
    t2.join()
    
for i in range(10):
    thread_main()
    print("{0} 번째 반복 이후 x = {1}".format(i, x))
```
